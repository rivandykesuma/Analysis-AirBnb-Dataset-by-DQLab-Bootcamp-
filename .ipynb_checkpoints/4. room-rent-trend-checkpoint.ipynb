```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn.objects as so
import seaborn as sns 
from scipy import stats
import numpy as np
import plotly.express as px
```

<H1>Room Rental Trend 01-01-2018 to 22-09-2022</H1>

```python
listing = pd.read_csv('data/listing.csv', index_col='Unnamed: 0')
neighbourhood = pd.read_csv('data/neighbourhood.csv', index_col='Unnamed: 0')
reviews = pd.read_csv('data/reviews.csv', index_col='Unnamed: 0')
```

```python
#mengidentifikasi Jenis Data 
listing.info()
print("=============================================")
neighbourhood.info()
print("=============================================")
reviews.info()
print("=============================================")
```

```python
#rename coloumn name 
reviews.rename(columns={'listing_id':'id'}, inplace=True)
```

```python
# merging Neighbourhood to listing 
df = listing.merge(neighbourhood, how='left', on='neighbourhood')
df.info() # All Data is NoNull 
```

```python
#Merging Listing to Reviews
df = reviews.merge(df, how='left', on='id')
df.info() # All Data is NoNull 
```

```python
df.head()
```

```python
#Converting date from object to datetime64
df['date'] = pd.to_datetime(df['date'])
df.info()
```

```python
#sort date
df = df.sort_values('date')
df
```

```python
#Get Data in columns date, host_name, room type
df = df[['date','host_name','room_type']]
df['year'] = df['date'].dt.to_period('Y') 
df = df[['year','host_name','room_type']]
df = df.groupby(['year', 'room_type']).count().reset_index()


```

```python
df.head()
```

```python
so.Plot(df, x='year', y='host_name', color='room_type').add(so.Line())
```
